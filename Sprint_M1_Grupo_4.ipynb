```python
class Producto(): # Defino la clase Producto
  def __init__(self,id,nombre,descripcion,cantidad,precio): # Inicializo sus atributos
    self.id = id
    self.nombre = nombre
    self.descripcion = descripcion
    self.cantidad = cantidad
    self.precio = precio

  def __str__(self): # Protocolo de cadena, me permite que sea más amigable de leer cuando use el comando print() o str() sobre este objeto
    return f"ID: {self.id}, Nombre: {self.nombre}, Descripción: {self.descripcion}, Cantidad: {self.cantidad}, Precio, {self.precio}"

auto = Producto(1, "Mazda2", "Auto Sedán", 3, 8000000) # Inicializo productos de ejemplo
moto = Producto(2, "Yamaha", "Moto 150CC", 4, 2000000)


class Inventario(): # Defino mi clase inventario
  def __init__(self):
    self.inventario = {}

  def listar(self): # Mi método listar revisa con un for los elementos de mi inventario y los va mostrando en pantalla, también calcula la cantidad total de productos y el valor total de estos
    print("Productos en el inventario")
    cantidad_total = 0
    valor_total = 0
    for producto in self.inventario.values():
      print(producto)
      cantidad_total += producto.cantidad
      valor_total += producto.precio
    print("Cantidad total: ", cantidad_total,"  Valor Total: ", valor_total)

  def agregar(self, producto): # Revisa primero si el id del producto existe, si no existe, permite agregarlo
    if producto.id in self.inventario:
      print("El ID del producto ya existe en el inventario")
    else:
      self.inventario[producto.id] = producto

  def eliminar(self, id_producto): # Revisa si el id se encuentra, si existe lo elimina
    if id_producto in self.inventario:
      del self.inventario[id_producto]
      print("eliminaste el producto con ID: ",
            id_producto)
    else:
      print("El id indicado no se puede eliminar porque no existe")

  def actualizar(self, id_producto, nombre=None, descripcion=None, cantidad=None, precio=None): # Permite actualizar los atributos del producto a excepción del id
    if id_producto in self.inventario:
      producto = self.inventario[id_producto]
      if nombre:
        producto.nombre = nombre
      if descripcion:
        producto.descripcion = descripcion
      if cantidad:
        producto.cantidad = cantidad
      if precio:
        producto.precio = precio
      else:
        print("El producto indicado no se puede actualizar, no existe en el inventario")

  def buscar(self, id_producto): # busca un producto por el id, si no lo encuentra, indica en pantalla que no existe
    if id_producto in self.inventario:
      print("El producto buscado es: ",self.inventario[id_producto])
    else:
      print("El producto no existe en el inventario")

  def buscar_id(self, id_producto): # Me permite buscar el id en el inventario y me devuelve el id buscado solo si lo encuentra, en otro caso me indica que el id no existe
    if id_producto in self.inventario:
      return id_producto
    else:
      print("No existe el id en el inventario")

inventario_1 = Inventario() # inicializo mi inventario
inventario_1.agregar(auto) # agrego los ejemplos
inventario_1.agregar(moto) # agrego los ejemplos



# Menú usuario
inicio = True
while inicio == True: # Mientras se mantenga verdadero me permite volver al menú principal

  try: # Me permite atajar las excepciones cuando le pido al usuario un int y me ingresa un string u otro elemento
   opcion = int(input('''Bienvenido al sistema de gestión de inventario, por favor ingresa tu opción
                      1. Listar productos
                      2. Agregar producto
                      3. Eliminar producto
                      4. Actualizar información del producto
                      5. Buscar producto por id
                      6. Salir
                      '''))

   if opcion == 1:

       inventario_1.listar()


   elif opcion == 2:

    agregar_id = int(input("ingresa el nuevo id de tu producto, debe ser un número entero positivo: "))
    aux = inventario_1.buscar_id(agregar_id) # mi variable aux me permite validar que el id ingresado no exista, para no tener duplicados

    if agregar_id > 0 and agregar_id != aux: # válida que el id sea positivo y que mi id nuevo no exista en la base
      print("Ingresa los datos de tu producto")
      nuevo_producto = Producto(agregar_id, input("Nombre: "), input("Descripción: "), int(input("Cantidad: ")),int(input("Precio: ")))
      inventario_1.agregar(nuevo_producto)
    else:
      print("No ingresaste un id válido o el id ingresado ya existe")

   elif opcion == 3: # simplemente compara la opción ingresada y ejecuta el metodo de mi inventario
     eliminar_id = int(input("ingresa el id del producto que deseas eliminar"))
     inventario_1.eliminar(eliminar_id)

   elif opcion == 4: # simplemente compara la opción ingresada y ejecuta el metodo de mi inventario
     actualizar_id = int(input("ingresa el id del producto que deseas actualizar"))
     inventario_1.actualizar(actualizar_id, input("Nombre: "), input("Descripción: "), input("Cantidad: "), input("Precio: "))

   elif opcion == 5: # simplemente compara la opción ingresada y ejecuta el metodo de mi inventario
     buscar_id = int(input("ingresa el id del producto que deseas buscar"))
     inventario_1.buscar(buscar_id)

   elif opcion == 6: # simplemente compara la opción ingresada y ejecuta el metodo de mi inventario
     inicio = False

   else: # si no ingresa un int válido, me sale este mensaje
     print("No ingresaste un comando válido, reintentalo denuevo")

  except ValueError: # me ataja las excepciones cuando le pido int(input) al usuario
    print("No ingresaste un comando válido")









```
